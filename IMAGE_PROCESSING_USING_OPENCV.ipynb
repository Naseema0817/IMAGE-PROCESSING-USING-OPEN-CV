Import necessary modules:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

LOAD THE IMAGE

```python
image = cv2.imread('Screenshot 2025-05-31 150635.png')  # Replace with your image name
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

ACCESS PIXCEL VALUES

```python
x, y = 50, 100  # adjust based on your image
(b, g, r) = image[y, x]
print(f"Pixel at ({x},{y}): B={b}, G={g}, R={r}")
```

Convert to grayscale:

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()

```

Apply Gaussian Blur

```python
blur = cv2.GaussianBlur(gray, (15, 15), 0)

plt.imshow(blur, cmap='gray')
plt.title("Blurred Image")
plt.axis("off")
plt.show()
```

Thresholding

```python
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap='gray')
plt.title("Thresholded Image")
plt.axis("off")
plt.show()
```

Edge Detection (Canny)

```python
edges = cv2.Canny(gray, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis("off")
plt.show()
```

Find and Draw Contours

```python
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
image_with_contours = image.copy()
cv2.drawContours(image_with_contours, contours, -1, (0, 255, 0), 2)

image_with_contours_rgb = cv2.cvtColor(image_with_contours, cv2.COLOR_BGR2RGB)
plt.imshow(image_with_contours_rgb)
plt.title("Contours")
plt.axis("off")
plt.show()
```

histogram(pixcel distribution)

```python
hist = cv2.calcHist([gray],[0],None,[256],[0,256])
plt.figure(figsize=(6,4))
plt.plot(hist, color='gray')
plt.title('Grayscale Histogram')
plt.xlabel('Pixel value')
plt.ylabel('Frequency')
plt.show()
```

Save Output Image

```python
cv2.imwrite("output_contours.jpg", image_with_contours)
files.download("output_contours.jpg")
```
